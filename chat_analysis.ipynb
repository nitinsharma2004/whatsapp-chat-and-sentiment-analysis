```python
import pandas as pd
import re
import matplotlib.pyplot as plt
```

```python
f = open('whatsapp_group.txt','r',encoding='utf-8')
```

```python
data = f.read()
```

```python
print(type(data))
```

```python
print(data)
```

```python
pattern = r'(\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s[AaPp][Mm]\s-\s)'
```

```python
matches = re.split(pattern, data)[1:]
matches
```

```python
cleaned_list = [item.replace('\u202f', ' ') for item in matches]
cleaned_list
```

```python
messages = []
for message in range(1,len(cleaned_list),2):
    messages.append(cleaned_list[message])
```

```python
messages
```

```python
len(messages)
```

```python
dates = re.findall(pattern,data)
dates = [item.replace('\u202f', ' ') for item in dates]
dates
```

```python
# Create DataFrame
df = pd.DataFrame({'user_message': messages, 'message_date': dates})
df.info()
```

```python
df['message_date'] = pd.to_datetime(df['message_date'], format='%d/%m/%Y, %I:%M %p - ')

df.rename(columns={'message_date': 'date'}, inplace=True)

df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
users = []
messages = []
for message in df['user_message']:
    entry = re.split('([\w\W]+?):\s',message)
    print
    if entry[1:]:
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group_notification')
        messages.append(entry[0])
df['user'] = users
df['message'] = messages
df.drop(columns = ['user_message'],inplace = True)
df.head()
```

#####  Sentiment analysis

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk

# Create SentimentIntensityAnalyzer
sentiments = SentimentIntensityAnalyzer()

# Assuming df is your DataFrame
positive = [sentiments.polarity_scores(i)["pos"] for i in df['message']]
negative = [sentiments.polarity_scores(i)["neg"] for i in df['message']]
neutral = [sentiments.polarity_scores(i)["neu"] for i in df['message']]
new_df = pd.DataFrame({'positive(+)': positive, 'negative(-)': negative,'neutral(N)':neutral})

new_df.head(5)
```

```python
average_sentiments = new_df.mean()

plt.figure(figsize=(8, 6))

average_sentiments.plot(kind='bar', color=['green', 'red', 'blue'])

plt.xlabel('Sentiment')
plt.ylabel('Average Value')
plt.title('Average Sentiment Distribution')
plt.xticks(rotation=0) 

plt.show()
```

```python

analyzer = SentimentIntensityAnalyzer()

def get_sentiment_score(text):
    sentiment = analyzer.polarity_scores(text)
    return sentiment['compound']

df['sentiment_score'] = df['message'].apply(get_sentiment_score)

def classify_sentiment(score):
    if score >= 0.05:
        return 'positive'
    elif score <= -0.05:
        return 'negative'
    else:
        return 'neutral'

df['sentiment'] = df['sentiment_score'].apply(classify_sentiment)
df
```
